## Data Cleaning and Preprocessing:-

as we checked.. there are some columns whichk have the missing values... We are going to fill that values using "fillna" function and giving it "0" value. 
we are not using "dropna" function as we are not drop that rows. 

```python
# Filling missing values:-

df["NumCompaniesWorked"].fillna(df['NumCompaniesWorked'].mode()[0], inplace=True)
df["TotalWorkingYears"].fillna(df['TotalWorkingYears'].mode()[0], inplace=True)
df["EnvironmentSatisfaction"].fillna(df['EnvironmentSatisfaction'].mode()[0], inplace=True)
df['JobSatisfaction'].fillna(df['JobSatisfaction'].mode()[0], inplace=True)
df['WorkLifeBalance'].fillna(df['WorkLifeBalance'].mode()[0], inplace=True)
```

Standard deviation(std) from the fields "EmployeeCount" and "StandardHours" are 0. Hence these fields doesnot add value, hence they can vbe removed.

### Categorical Values:-

We are going to encoding categorical variables into numerical values. Machine learning algorithms generally require numerical input, so categorical data needs to be converred to a numerical format.

```python
# Importing "LabelEncoder" library for categorical Vlaues:-

from sklearn.preprocessing import LabelEncoder
```

```python
label_encoders = {}
categorical_columns = ['Attrition', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus']

for column in categorical_columns:
    label_encoders[column] = LabelEncoder()
    df[column] = label_encoders[column].fit_transform(df[column])
```

##### Analyzing the relationship between "Attrition" and "Jobrole" to understand which job roles have higher attrition rates.

```python
# Calculating the total number of employees in each job role:-
total_emp_by_Role = df['JobRole'].value_counts()

# Calcualting the number of employees who left in each job role:-
attrition_by_role = df[df['Attrition'] == 1]['JobRole'].value_counts()

# Calculating the Attrition rate for each job role:-
attrition_rate_by_role = (attrition_by_role / total_emp_by_Role) * 100

# Converting the result to a DataFrame for better readability:-
attrition_rate_by_role_df = attrition_by_role.reset_index()
attrition_rate_by_role_df.columns = ['JobRole', "AttritionRate"]

# Display the attrition rate for each job role:-
print(attrition_rate_by_role_df)

# Let's sort the Data of "attrition_rate_by_role_df"..
attrition_rate_by_role_df = attrition_rate_by_role_df.sort_values(by = 'AttritionRate', ascending=False)
attrition_rate_by_role_df
```

#### Plotting the attrition rate for each "job role":-


```python
# Retrive the labelEncoder for the "JobRole" column:-
job_role_encoder = label_encoders["JobRole"]

# Creating a dictionary to map numerical values to job roles:-
job_role_mapping = {index: role for index, role in enumerate(job_role_encoder.classes_)}

for num, role in job_role_mapping.items():
    print(f"{num} means {role}")
```

## Model Building and Evalualtion:-

```python
# import randomforest and train_test_split libraries:-

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Define featiures and target variable.
X = df.drop('Attrition', axis=1)
y = df['Attrition']

# Split the data into training and testing sets:-
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train the model:-
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Make predictions:-
y_pred = model.predict(X_test)

# Evaluating the Model.
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Feature Importance:-
plt.figure(figsize=(10, 8))
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]
features = X.columns


plt.title('Feature Importances')
plt.bar(range(X.shape[1]), importances[indices], align='center')
plt.xticks(range(X.shape[1]), features[indices], rotation=90)
plt.tight_layout()
plt.show()
```

#### Conclusion from the Feature Importance Plot:-

* "MonthlyIncome" has the highest importance score, indicating it is the most significant predictor of employee attrition in the dataset.
* "Age" is also a crucial factor. Older employees might be less likely to leave compared to younger ones.
* Higher "job satisfaction" is associated with lower attrition, making it a vital feature for predicting employee turnover.
